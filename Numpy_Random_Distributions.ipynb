```python
# Normal Distribution (Gaussian Distribution)

# It Uses the random.normal() method to get a Normal Data Distribution.
# It has three parameters:
#loc - (Mean) where the peak of the bell exists.
# scale - (Standard Deviation) how flat the graph distribution should be.
# size - The shape of the returned array.

a = np.random.normal(size=(2, 2))

print(a)
```

```python
# Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2:

a = random.normal(loc=1, scale=2, size=(2, 3))

print(a)
```

```python
# Visualization of normal distribution

sns.displot(random.normal(size=1500), kind="kde")

plt.show()
```

```python
# Binomal Distribution

# Binomial Distribution is a Discrete Distribution.
# It describes the outcome of binary scenarios, e.g. toss of a coin, it will either
#be head or tails.
# It has three parameters:
# n - number of trials.
# p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).
# size - The shape of the returned array.
```

```python
# Example of Binormal distribution

a = random.binomial(n= 10, p=0.5, size=5)

print(a)
```

```python
# Visualizing Binormal Distribution

sns.displot(random.binomial(n =10, p =0.5, size =1500))
plt.show()
```

```python
# Combinning normal and binomial Distribution

data = {
    "normal": np.random.normal(loc=50, scale=5, size=1500),
    "binomial": np.random.binomial(n=100, p=0.5, size=1500)
}

sns.displot(data, kind="kde")

plt.show()
```

```python
#Poisson Distribution

# Poisson Distribution is a Discrete Distribution which estimates how many times an event can happen in a specified time.
# It has two parameters:
# lam - rate or known number of occurrences e.g. 4 for below problem.
# size - The shape of the returned array.

a = np.random.poisson(lam=4, size=15)

print(a)
```

```python
# Plotting a poisson Distribution
sns.displot(random.poisson(lam=4, size=1500))

plt.show()
```

```python
# plotting Normal and Poisson Distribution together

Data = {
    "normal":    random.normal(loc=50, scale=7, size=1000),
    "poisson":   random.poisson(lam=50, size=1000)
}

sns.displot(Data, kind="kde")

plt.show()
```

```python
# Plotting Poisson and Binimial Distributions Together

data = {
    "binomial": random.binomial(n=1000, p=0.01, size=1000),
    "poisson": random.poisson(lam=10, size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

```python
# Uniform Distribution

# The normal distribution is used to describe probability where every event has equal chances of occuring.E.g. Generation of random numbers.

# It has three parameters:

# low - lower bound - default 0 .0.

# high - upper bound - default 1.0.

# size - The shape of the returned array.
```

```python
# Generating a 2x4 uniform Distribution

a = random.uniform(size=(2, 3))

print(a)
```

```python
# visualizing a Uniform Distribution

sns.displot(random.uniform(size=1000), kind="kde")

plt.show()
```

```python
# Logistic Distribution

#Logistic Distribution is used to describe growth.

#Used extensively in machine learning in logistic regression, neural networks etc.

#It has three parameters:

#loc - mean, where the peak is. Default 0.

#scale - standard deviation, the flatness of distribution. Default 1.

#size - The shape of the returned array.

```

```python
a = random.logistic(loc=1, scale=5, size=(4, 2))

print(a)
```

```python
# Visualization of Logistic Distribution

sns.displot(random.logistic(size=1000), kind="kde")

plt.show()
```

```python
# Plotting the normal and Logistic distribution together

data = {
    "normal": random.normal(scale=2, size=1000),
    "logistic": random.logistic(size=1000)
}

sns.displot(data, kind="kde")

plt.show()
```

```python
# Multinomial Distribution

# Multinomial distribution is a generalization of binomial distribution.

# It describes outcomes of multi-nomial scenarios unlike binomial where scenarios must be only one of two. e.g. Blood type of a population, dice roll outcome.

# It has three parameters:

# n - number of times to run the experiment.

# pvals - list of probabilties of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll).

# size - The shape of the returned array.
```

```python
# Multi distribution from a dice roll

a = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

print(a)
```

```python
# Exponential Distribution

# Exponential distribution is used for describing time till next event e.g. failure/success etc.

# It has two parameters:

# scale - inverse of rate ( see lam in poisson distribution ) defaults to 1.0.

# size - The shape of the returned array.
```

```python
# Draw out a sample for exponential distribution with 2.0 scale with 2x4 size:

a = random.exponential(scale=2, size=(2, 4))

print(a)
```

```python
#Visualization of Exponential Distribution

sns.displot(random.exponential(size=1500), kind="kde")

plt.show()
```

```python
# Chi Square Distribution

#Chi Square distribution is used as a basis to verify the hypothesis.

#It has two parameters:

# df - (degree of freedom).

# size - The shape of the returned array.
```

```python
# Draw out a sample for chi squared distribution with degree of freedom 2 with size 2x3:

a = random.chisquare(df=2, size=(2, 4))

print(a)
```

```python
# Visualization of Chi Square Distribution

sns.displot(random.chisquare(df=1, size=1500), kind="kde")

plt.show()
```

```python
# Rayleigh Distribution
# Rayleigh distribution is used in signal processing.

# It has two parameters:

# scale - (standard deviation) which decides how flat the distribution will be. Default is 1.0).

# size - The shape of the returned array.
```

```python
#Draw out a sample for rayleigh distribution with scale of 2 with size 2x4:

a = random.rayleigh(scale=2, size=(2, 4))

print(a)
```

```python
sns.displot(random.rayleigh(size= 1500),kind = 'kde')
plt.show()
```

```python
# Pareto Distribution
# A distribution following Pareto's law i.e. 80-20 distribution (20% factors cause 80% outcome).

# It has two parameter:

# a - shape parameter.

# size - The shape of the returned array.
```

```python
# Draw out a sample for pareto distribution with shape of 2 with size 2x4:

from numpy import random

a = random.pareto(a=2, size=(2, 4))

print(a)
```

```python
sns.displot(random.pareto(a =2, size = 1500))
plt.show()
```

```python
# Zipf Distribution

# Zipf distributions are used to sample data based on zipf's law.
# Zipf's Law: In a collection, the nth common term is 1/n times of the most common term.
E.g. the 10th most common word in English occurs nearly 1/10 times as often as the most common word.

# It has two parameters:
#a - distribution parameter.
#size - The shape of the returned array.
```

```python
# Draw out a sample for zipf distribution with distribution parameter 2 with size 2x4:


a = random.zipf(a=2, size=(2, 4))

print(a)
```

```python
# Visualization of Zipf Distribution
#Sample 1500 points but plotting only ones with value < 50 for more meaningful chart.

d =(random.zipf(a = 2, size = 1500))
sns.displot(d[d<50])
plt.show()
```
